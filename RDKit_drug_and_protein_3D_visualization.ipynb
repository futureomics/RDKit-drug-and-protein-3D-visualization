```python
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol

# Create a molecule from SMILES
smiles = "CC(=O)Oc1ccccc1C(O)=O"
mol = Chem.MolFromSmiles(smiles)

# Generate 3D coordinates
AllChem.EmbedMolecule(mol)
AllChem.UFFOptimizeMolecule(mol)

# Convert to a 3D structure format
block = Chem.MolToMolBlock(mol)

# Visualize with py3Dmol
viewer = py3Dmol.view(width=400, height=300)
viewer.addModel(block, 'mol')
#viewer.setStyle({'stick': {}})
viewer.setStyle({'sphere': {}})
viewer.setBackgroundColor('white')
viewer.zoomTo()
viewer.show()
```

```python
# Visualize with py3Dmol
viewer = py3Dmol.view(width=600, height=400)
viewer.addModel(block, 'mol')

# Set different styles and colors
viewer.setStyle({'stick': {}})
viewer.setStyle({'sphere': {'colorscheme': 'cyanCarbon'}})
viewer.addSurface(py3Dmol.SAS, {'opacity': 0.3, 'color': 'yellow'})

# Change background color
viewer.setBackgroundColor('lightgray')

# Zoom to the entire molecule
viewer.zoomTo()

# Show the visualization
viewer.show()
```

```python
import py3Dmol

# Load protein structure from a PDB file
pdb_file = "/content/4uya.pdb"  # Replace with your PDB file path
with open(pdb_file) as f:
    protein_data = f.read()

# Create the viewer
viewer = py3Dmol.view(width=800, height=400)
viewer.addModel(protein_data, 'pdb')

# Set cartoon representation
viewer.setStyle({'cartoon': {'color': 'spectrum'}})  # Color by spectrum

# Add ball-and-stick representation for side chains
viewer.addStyle({'stick': {}, 'sphere': {'scale': 0.3}})  # Sticks for bonds, spheres for atoms

# Highlight specific residues in ball-and-stick style
viewer.setStyle({'resi': '45'}, {'stick': {'colorscheme': 'purpleCarbon'}})  # Highlight residue 45
viewer.setStyle({'resi': '50'}, {'stick': {'colorscheme': 'orangeCarbon'}})  # Highlight residue 50

# Optional: Add surface representation to visualize interaction areas
viewer.addSurface(py3Dmol.SAS, {'opacity': 0.3, 'color': 'lightgray'})

# Annotate important residues
viewer.addLabel('Active Site', {'fontSize': 16, 'position': {'resi': '50'}})

# Set background and show the viewer
viewer.setBackgroundColor('white')
viewer.zoomTo()
viewer.show()
```
